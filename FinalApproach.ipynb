```python
import pandas as pd
import numpy as np
trainingSet = pd.read_csv("./data/train.csv")
testingSet = pd.read_csv("./data/test.csv")
```

```python
predictionSet = pd.merge(trainingSet, testingSet, left_on='Id', right_on='Id')
#print(predictionSet.columns)

predictionSet = predictionSet.drop(columns=['Score_x'])
predictionSet = predictionSet.rename(columns={'Score_y': 'Score'})

#print(predictionSet.columns)
# predictionSet.to_csv("./data/prediction.csv", index=False)

X_train = trainingSet[trainingSet['Score'].notnull()]
#df = X_train.drop(columns=['HelpfulnessNumerator', 'HelpfulnessDenominator', 'ProductId', 'UserId', 'Time', 'Summary'])
df=X_train.dropna(axis=0)
```

```python
products = np.unique(trainingSet.ProductId.values)
```

```python
df.head()
```

```python
product_count = {}
for i,s in enumerate(products):
    product_count[s] = 0
product_sum = {}
for i,s in enumerate(products):
    product_sum[s] = 0

for i,s in enumerate(df["ProductId"]):
    product_count[s] += 1
for i,s in enumerate(df["ProductId"]):
    product_sum[s] += df.Score.values[i]
```

```python
for key in product_sum:
    product_sum[key] = np.round_(product_sum[key]/max(1,product_count[key]))
product_sum
```

```python
users = np.unique(trainingSet.UserId.values)
```

```python
user_count = {}
for i,s in enumerate(users):
    user_count[s] = 0
    
user_sum = {}
for i,s in enumerate(users):
    user_sum[s] = 0
    
for i,s in enumerate(df["UserId"]):
    user_count[s] += 1
for i,s in enumerate(df["UserId"]):
    user_sum[s] += df.Score.values[i]
```

```python
for key in user_sum:
    user_sum[key] = np.round_(user_sum[key]/max(1,user_count[key]))
user_sum
```

```python
df_copy = trainingSet.drop(columns=["Summary", "Text"])
df_copy["UserAverage"] = np.zeros(len(df_copy), dtype=np.float64)
df_copy["MovieAverage"] = np.zeros(len(df_copy), dtype=np.float64)
for i in range(len(df_copy)):
    df_copy.UserAverage.values[i] = user_sum[df_copy.UserId[i]]
for i in range(len(df_copy)):
    df_copy.MovieAverage.values[i] = product_sum[df_copy.ProductId[i]]

df_copy.head()
```
